```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Definir as constantes
H0 = 70  # Parâmetro de Hubble em km/s/Mpc
H0_si = H0 * 1000 / (3.086e22)  # Converter para s^-1
Omega_m = 0.3
Omega_r = 9.24e-5
Omega_g = 0.100
Omega_suck = 0.599
n = 3.495

# Definir a equação de Friedmann
def friedmann(a, t, Omega_m, Omega_r, Omega_g, Omega_suck, n):
    return H0_si * np.sqrt(Omega_m * a**(-3) + Omega_r * a**(-4) + Omega_g * a**(-3) + Omega_suck * a**(-n))

# Resolver a equação de Friedmann
a0 = 1e-4
t = np.linspace(0, 13.8e9 * 365.25 * 24 * 3600, 1000)  # Tempo em segundos
a_opt = odeint(friedmann, a0, t, args=(Omega_m, Omega_r, Omega_g, Omega_suck, n))

# Função de densidade de energia (logarítmica)
def rho_suck(a, Omega_suck, n):
    return Omega_suck * np.log(n * a + 1)

# Calcular a densidade de energia da sucção ao longo do tempo
rho_suck_evol = rho_suck(a_opt, Omega_suck, n)

# Plotar a evolução das densidades de energia
plt.figure(figsize=(10, 6))
plt.plot(t / (365.25 * 24 * 3600 * 1e9), Omega_m * a_opt[:,0]**(-3), label='Matéria')
plt.plot(t / (365.25 * 24 * 3600 * 1e9), Omega_r * a_opt[:,0]**(-4), label='Radiação')
plt.plot(t / (365.25 * 24 * 3600 * 1e9), rho_suck_evol, label='Sucção')
plt.xlabel('Tempo (bilhões de anos)')
plt.ylabel('Densidade de Energia')
plt.title('Evolução das Densidades de Energia')
plt.legend()
plt.yscale('log')
plt.grid(True)
plt.show()
```
